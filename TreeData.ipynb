```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
trees = pd.read_csv("PittsburghTrees.csv", low_memory=False)
num_trees = trees["neighborhood"].value_counts()
print(num_trees)
```

Organizing the data by number of trees in each neighborhood. There are a lot of trees. 

```python
num_trees.plot.bar()
```

Here is a bar graph of the number of trees in each neighborhood. There is a lot going on here, it is not easy to interpret. 

```python
trees_query_mask = num_trees > 500
lots_of_trees = num_trees[trees_query_mask]
print(lots_of_trees)
```

To make it easier to visualize, a querey mask is made. It only includes neighborhoods with over 500 trees. 

```python
lots_of_trees.plot.bar()
```

Visualization of the data of neighborhoods with over 500 trees. This is useful because the neighborhoods with more trees are better candidates for the best neighborhood for a kid. This is much easier to understand than all of the data put into one graph. Because there aren't many trees in them, the neighborhoods not included in this plot are not good candidates for best neighborhood for a kid.
